```python
import pandas as ps
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y)

plt.title('BEGINNER', fontdict={'fontname':'comic sans MS', 'fontsize':20})
plt.xlabel('x Axis')
plt.ylabel('y Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,7,5,10])

plt.show()
```

```python
labels = ['java','python','c']
values = [1,4,2]

bars=plt.bar(labels,values)

patterns = ['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.title('Programming Languages')

plt.figure(figsize=(6,4))

plt.show()
```

```python
subject = ['tamil', 'english', 'maths',
		'science', 'science', 'social']

data = [23, 17, 35, 29, 12, 41]
fig = plt.figure(figsize =(10, 5))
plt.pie(data, labels = subject)


plt.show()
```
